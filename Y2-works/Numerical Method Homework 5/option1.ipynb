```python
# homework 5.2 root-finding methods (for optimization)
# f1(x) = x^3-6x^2+4x+12
# f2(x) = x^3-4x^2-6x+5
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import sympy
```

```python
# define x and f symbolically
x = sympy.Symbol('x')
fx = x**3 - 6*(x**2)+(4*x)+12
dfx = fx.diff(x)
fx = sympy.lambdify(x,fx)
dfx = sympy.lambdify(x,dfx)
```

```python
# visualize fx and dfx
fig = plt.figure(figsize=(10,6))
x = np.arange(-2,6,0.1)
y_fx = fx(x)
y_dfx = dfx(x)
plt.plot(x, y_fx, c = 'b', label = 'fx')
plt.plot(x, y_dfx, c = 'r', label = 'dfx')
plt.grid(True, which='both')
plt.axhline(y = 0, color='k')
plt.legend()
plt.show()
```
